# [ 조건문 if ]
-⭕ 조건은 소괄호 사용 ()

-⭕ 실행할 내용은 중괄호 사용 {}


[ 조건문 ]
1. 기본형 : if 조건:  / if (조건) {...}
2. 위의 조건이 아니면 : else / else {...}
3. 1,2 조건이 둘다 아니면 : elfe 조건: / else if (조건) {...}

```python
# 조건문 코드

public class IfExample {
    public static void main(String[] args) {
        int age = 18;

        // 자바의 조건문 공식: if (조건) { 실행 }
        if (age >= 20) {
            System.out.println("성인입니다. 맥주를 살 수 있어요!");
        }
        else if (age >= 14) {
            System.out.println("청소년입니다. 학생 할인이 가능해요.");
        }
        else {
            System.out.println("어린이입니다. 보호자와 동반하세요.");
        }
    }
}
```

```python
# [ 반복문 while ]

public class WhileExample {
    public static void main(String[] args) {
        int count = 0;

        //count가 3보다 작을 동안 계속 반복해!
        while (count < 3) {
            System.out.println("현재 숫자는: " + count);
            count++; // 이 줄이 없으면 무한 반복에 빠지니 주의!
        }

        System.out.println("반복 종료!");
    }
}
```

```python
# 예제) 1부터 5까지 더하기 (합 구하기)

public class whileSum {
    public static void main(String[] args){

        int i = i;       # i는 현재 숫자
        int sum = 0;     # sum 누적 저장 공간

        while (i <= 5) {
            sum += i;
            i++;
        }
        System.out.println("합: " + sum);
    }
}
```

# [ 반복문 제어: break / continue]

1. break (멈춰)
- 반복문을 빠져나오는 용도
- 조건 상관없이 반복 종료

2. continue (이번만 넘겨)
- 이번 반복만 스킵
- 반복은 계속 유지함
- 특정 순간만 skip

```python
# 예제) break

public class BreakExample {
    public static void main(String[] args) {

        int i = 1;

        while (i <= 10) {

            if (i == 5) {
                break; // 반복 종료
            }

            System.out.println(i);
            i++;
        }

        System.out.println("반복 끝");
    }
}
```

```python
# 예제) continue

public class ContinueExample {
    public static void main(String[] args) {

        int i = 0;

        while (i < 5) {
            i++;

            if (i == 3) {
                continue; // 3만 출력 안 함
            }

            System.out.println(i);
        }
    }
}
```
