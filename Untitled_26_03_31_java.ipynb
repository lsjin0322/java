# [ 조건문 if ]
-⭕ 조건은 소괄호 사용 ()

-⭕ 실행할 내용은 중괄호 사용 {}


[ 조건문 ]
1. 기본형 : if 조건:  / if (조건) {...}
2. 위의 조건이 아니면 : else / else {...}
3. 1,2 조건이 둘다 아니면 : elfe 조건: / else if (조건) {...}

```python
# 조건문 코드

public class IfExample {
    public static void main(String[] args) {
        int age = 18;

        // 자바의 조건문 공식: if (조건) { 실행 }
        if (age >= 20) {
            System.out.println("성인입니다. 맥주를 살 수 있어요!");
        }
        else if (age >= 14) {
            System.out.println("청소년입니다. 학생 할인이 가능해요.");
        }
        else {
            System.out.println("어린이입니다. 보호자와 동반하세요.");
        }
    }
}
```

```python
public class StringCompare {
    public static void main(String[] args) {
        String myName = "이수진";       # 입력

        if (myName.equals("이수진")){   # ❗글자 비교는 == 대신 .equals()를 쓴다!
            System.out.println ("제 이름이 맞네요!")    # 실행할 내용
        }
    }
}

# point
# 1.자바의 if 뒤에는 반드시 소괄호 ( ) 가 따라온다!

# 2. 실행할 내용은 중괄호 { } 라는 울타리 안에 가둔다!

# 3. 파이썬의 elif는 자바에서 else if  라고 길게 풀어 쓴다!
```

# [ 반복문 ]

- for (시작; 조건; 변화) {...} 처럼 아주 구체적으로 명령해야함
- 정확한 횟수가 정해져 있을 때 사용함.  
    (ex) 1번부터 100번까지 반복해
- 반복할 내용에도 중괄호 {}를 넣어준다.

-▶ int i = 0; ( 시작/ 0부터 시작할게! )


-▶ i < 10; ( 10보다 작을 떄까지만 해!)

-▶ i++ ( 한 번 끝날때마다 숫자를 1씩 키워! )

    i = i + 1

-▶ 1-- ( 한 번 끝날때마다 숫자를 1씩 빼 ! )

-▶ x -= 1 ( x는 x-1이라는 뜻/ 사용가능기호 : + ,/ ,*, % )



```python
# [ 반복문 if ]
public class ForExample {
    public static void main(String[] args) {

    //"안녕하세요"를 5번 출력하고 싶을 때
    // (시작값; 끝조건; 증가값 (변화))

        for(int i = 1; i <= 5; i++) {    # i를 1로 선언한다. , i가 5보다 작거나 같으면 i에 1씩 더하기
            System.out.println(i + "번째 안녕하세요!");   # i에 나오는 숫자표시 후 문자도 프린트
        }
    }
}
```

# [ 반복문 while ] ❌ 무한루프 주의

1. 반복 조건이 T이면 반복, F면 종료
2. 반복 조건이 없으면 컴파일러 오류
3. 처음부터 반복조건을 통과한 후 작업문 수행
4. 조건만 있으면 사용

    (ex) 김영수를 찾아라

[ 식 ]

while (조건) {...}


```python
# [ 반복문 while ]

public class WhileExample {
    public static void main(String[] args) {
        int count = 0;

        //count가 3보다 작을 동안 계속 반복해!
        while (count < 3) {
            System.out.println("현재 숫자는: " + count);
            count++; // 이 줄이 없으면 무한 반복에 빠지니 주의!
        }

        System.out.println("반복 종료!");
    }
}
```
